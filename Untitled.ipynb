```python
from PIL import Image
img = Image.open("850c3min_15000_1.jpg").convert('L')

size = img.size

img2 = Image.new('L', size)

for x in range(size[0]):
    for y in range(size[1]):
        c = img.getpixel((x, y))
        
        if c > 50:
            c = 1
        else:
            c = 0

        img2.putpixel((x,y),c)


img2.show()
```

```python
from PIL import Image, ImageFilter, ImageChops
img= Image.open("850c3min_15000_1.jpg")
size = img.size

img2 = Image.new('L', size, 255)
r, g, b = img.split()

src_color = (0, 0, 0)

_r = r.point(lambda _: 1 if _ == src_color[0] else 0, mode="1")
_g = g.point(lambda _: 1 if _ == src_color[1] else 0, mode="1")
_b = b.point(lambda _: 1 if _ == src_color[2] else 0, mode="1")

mask = ImageChops.logical_and(_r, _g)
mask = ImageChops.logical_and(mask, _b)

filtered = mask.filter(ImageFilter.CONTOUR)

mask.show()
```

```python
%matplotlib inline
import numpy as np
from scipy import ndimage, misc
import matplotlib.pyplot as plt
from PIL import Image


img = np.array(Image.open('sample.png').convert('L'), 'f')

img = 1 * (img > 128)

open_img = ndimage.binary_opening(img)
close_img = ndimage.binary_closing(open_img)
#plt.subplot(121)
plt.imshow(close_img, cmap='Greys_r')

lx, ly = close_img.shape
l = 5100
plt.contour(close_img[:lx, :ly], [0.5], linewidths=0.2, colors='r')
plt.axis('off')
plt.savefig("out.png",format = 'png', dpi=300)
```
